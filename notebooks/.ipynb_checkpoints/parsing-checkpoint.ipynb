```python
import torch
from torch_struct import SentCFG

NEG_INF = -1e9

# -------------------------
# 1. CNF-style grammar
# -------------------------
non_terminals = [
    "Gene", "Regulatory", "Coding", "Promoter", "Enhancer",
    "Exon", "Intron", "B", "X1", "X2", "X3", "X4", "X5"
]
terminals = ["A", "T", "G", "C"]

nt_to_idx = {nt: i for i, nt in enumerate(non_terminals)}
t_to_idx = {t: i for i, t in enumerate(terminals)}

NT = len(non_terminals)
T = len(terminals)
S = NT + T

def sym_idx(sym):
    if sym in nt_to_idx:
        return nt_to_idx[sym]
    return NT + t_to_idx[sym]

# rules[A, B, C] = score for A -> B C
rules = torch.full((1, NT, S, S), NEG_INF)

def add_rule(lhs, rhs1, rhs2, score=0.0):
    rules[0, nt_to_idx[lhs], sym_idx(rhs1), sym_idx(rhs2)] = score

# Gene -> Regulatory Coding | Exon IntronLike
add_rule("Gene", "Regulatory", "Coding")
add_rule("Gene", "Exon", "Exon")

# Regulatory -> Promoter Enhancer | Promoter Exon
add_rule("Regulatory", "Promoter", "Enhancer")
add_rule("Regulatory", "Promoter", "Exon")

# Coding -> Exon Intron | Exon Exon
add_rule("Coding", "Exon", "Intron")
add_rule("Coding", "Exon", "Exon")

# Exon -> B B
add_rule("Exon", "B", "B")

# Intron -> B B
add_rule("Intron", "B", "B")

# Binarized TATA:
# Promoter -> T X1
# X1 -> A X2
# X2 -> T A
add_rule("Promoter", "T", "X1")
add_rule("X1", "A", "X2")
add_rule("X2", "T", "A")

# Binarized GGGCGG:
# Enhancer -> G X3
# X3 -> G X4
# X4 -> G X5
# X5 -> C X2g
# reuse X2? better separate:
non_terminals_extra = False
```

```python
import torch
from torch_struct import SentCFG
import graphviz  # You must: pip install graphviz

NEG_INF = -1e9

# --------------------------------------------------
# 1. Grammar Setup
# --------------------------------------------------
non_terminals = ["Gene", "Promoter", "Coding", "P1", "P2"]
terminals = ["A", "T", "G", "C"]

nt_to_idx = {nt: i for i, nt in enumerate(non_terminals)}
t_to_idx = {t: i for i, t in enumerate(terminals)}

NT = len(non_terminals)
T = len(terminals)
S = NT + T 

def sym_name(idx):
    if idx < NT: return non_terminals[idx]
    return terminals[idx - NT]

def sym_idx(sym):
    if sym in nt_to_idx: return nt_to_idx[sym]
    return NT + t_to_idx[sym]

rules = torch.full((1, NT, S, S), NEG_INF)
def add_rule(lhs, rhs1, rhs2, score=0.0):
    rules[0, nt_to_idx[lhs], sym_idx(rhs1), sym_idx(rhs2)] = score

add_rule("Gene", "Promoter", "Coding")
add_rule("Promoter", "T", "P1")
add_rule("P1", "A", "P2")
add_rule("P2", "T", "A")
add_rule("Coding", "G", "G")

root = torch.full((1, NT), NEG_INF)
root[0, nt_to_idx["Gene"]] = 0.0

seq = "TATAGG"
N = len(seq)
terms = torch.full((1, N, T), NEG_INF)
for i, ch in enumerate(seq):
    terms[0, i, t_to_idx[ch]] = 0.0

# --------------------------------------------------
# 2. Run Parsing
# --------------------------------------------------
dist = SentCFG((terms, rules, root), lengths=torch.tensor([N]))
arg = dist.argmax

rule_chart = arg[1][0] 
root_chart = arg[2][0]

# --------------------------------------------------
# 3. Recursive Reconstruction (Standardized Format)
# --------------------------------------------------
active_rules = (rule_chart > 0.5).nonzero(as_tuple=False)
split_map = {item[0].item(): (item[1].item(), item[2].item()) for item in active_rules}

def build_tree_data(i, j, sym_i):
    """
    Standardizes the tree into dictionaries: {"name": "X", "children": [...]}
    """
    name = sym_name(sym_i)
    
    # Base Case: Single character/terminal
    if j == i + 1:
        # We always want the symbol (e.g., T) to point to the actual base (e.g., T)
        # to distinguish Non-Terminals from Terminal leaves.
        return {
            "name": name, 
            "children": [{"name": seq[i], "children": []}]
        }
    
    # Recursive Case: Find the split point that belongs to this span
    for k_idx, (left_sym, right_sym) in split_map.items():
        k = k_idx + 1
        
        if i < k < j:
            return {
                "name": name, 
                "children": [
                    build_tree_data(i, k, left_sym),
                    build_tree_data(k, j, right_sym)
                ]
            }

    return {"name": name, "children": []}

# Identify starting Non-Terminal
start_nt = (root_chart > 0.5).nonzero(as_tuple=False)[0].item()
tree_data = build_tree_data(0, N, start_nt)

# --------------------------------------------------
# 4. Visualization Functions
# --------------------------------------------------

def visualize_pretty(node, prefix="", is_last=True):
    """ Console tree-style printer """
    marker = "└── " if is_last else "├── "
    print(prefix + marker + node["name"])
    new_prefix = prefix + ("    " if is_last else "│   ")
    
    child_count = len(node["children"])
    for i, child in enumerate(node["children"]):
        visualize_pretty(child, new_prefix, i == child_count - 1)

def visualize_graphviz(tree_root, filename='gene_tree'):
    """ Generates a Graphviz PNG visualization """
    dot = graphviz.Digraph(comment='Gene AST')
    dot.attr(nodesep='0.5', ranksep='0.5')
    
    # Clean visual style for non-terminal nodes (blue text, no box)
    dot.node_attr.update(shape='none', fontname='helvetica', fontcolor='blue')

    counter = 0

    def add_nodes(parent_id, node):
        nonlocal counter
        node_id = str(counter)
        counter += 1
        
        # Determine label and style
        label = node["name"]
        
        if not node["children"]:
            # This is a leaf node (the DNA bases), color it green and bold
            dot.node(node_id, label, fontcolor='darkgreen', fontname='helvetica-bold')
        else:
            dot.node(node_id, label)
        
        if parent_id:
            dot.edge(parent_id, node_id)
        
        for child in node["children"]:
            add_nodes(node_id, child)

    add_nodes(None, tree_root)
    
    # Render as PNG
    try:
        dot.render(filename, format='png', cleanup=True)
        print(f"Graphviz PNG saved as '{filename}.png'!")
    except graphviz.backend.ExecutableNotFound:
        print("Error: Graphviz executable not found. Make sure Graphviz is installed and in your PATH.")

# --------------------------------------------------
# 5. Output
# --------------------------------------------------
print("\n[1] CONSOLE VISUALIZATION")
print("="*30)
visualize_pretty(tree_data)

print("\n[2] GENERATING GRAPHVIZ PNG")
print("="*30)
visualize_graphviz(tree_data, filename='gene_structure')
```
